# 복습 문제-함수

```python
def calculate_average(numbers):
    total=0
    for n in numbers:   
        total+=n
    return total/len(numbers)
my_scores=[80,90,70,100]
avg=calculate_average(my_scores)        
print("평균 :", avg)
```

복습 문제-종합

Q)
3명의 학생 이름과 점수가 있고 점수에 따라 등급을 매기는 함수를 생성하고 로그를 출력합니다.

EX)로그 출력
-A 학생은 A등급입니다.
-B 학생은 C등급입니다.
-C 학생은 B등급입니다.

```python
def get_grade(score):
    if score >= 90:
        return "A"
    elif score >= 80:
        return "B"
    elif score >= 70:
        return "C"
    else:
        return "D"


students = ["A", "B", "C"]
student_scores = [95, 73, 58]

for i in range(len(students)):
    name = students[i]
    score = student_scores[i]
    grade = get_grade(score)

    print(name, "학생은", grade, "등급입니다.")

```

# 모델 학습과 예측 실습

## 모델이란?

> 정의 : 데이터를 보고 규칙을 학스한 뒤,  
> 새로운 데이터가 들어오면 결과를 예측하는 프로그램

쉬운 비유:
- 모델=학생
- 학생(모델)에게 문제와 답,개념(데이터)를 주고 공부를 시킵니다.
- 공부가 끝나면, 새로운 문제(데이터)를 풀게 합니다.

# 학습(fit)이란?

> 정의 : 모델이 입력값(x)과 정답(y)를 보고  
> 둘 사이의 패턴이나 규칙을 찾는 과정

- `model.fit(x,y)` 을 사용하면 학습이 시작됩니다.
- x(입력값) 와 y(정답)을 함께 넣어줘야 합니다.
- 모델은 학습이 끝나고 규칙을 기억하고 있게 됩니다.

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier

data={
    "study_hours": [1,2,3,3,4,4,5,6,7,8],
    "attendance": [50,55,60,80,60,75,75,80,90,95],
    "pass_fail": [0,0,0,1,0,1,1,1,1,1]
}
df=pd.DataFrame(data)

# 입력값과 정답 분리
x=df[["study_hours", "attendance"]]
y=df["pass_fail"]

print("데이터 준비 완료")
print(f"학생 수: {len(df)}명") # 학생 수: %%명
```

```python
# 모델 만들기

model = DecisionTreeClassifier(random_state=42)

model.fit(x, y)

print("모델 학습 완료")
```

```python
# 새 학생 1명을 예측해보기
# 주당 4시간 공부, 출석률을 80%

new_student=pd.DataFrame([[4,80]], columns=["study_hours", "attendance"])
prediction=model.predict(new_student)
print("공부 4시간, 출석 80%인 학생:", prediction)


if prediction[0]==1:
    print("모델 판단: 합격")
else:
    print("모델 판단 : 불합격")
```

```python
# 여러 학생 예측하기

new_students=pd.DataFrame([
    [4,80],
    [3,65],
    [5,80],
    [7,95]
    ], columns=["study_hours", "attendance"])  # 새 학생들 정보 표로 저장 및 열 이름 설정

results=model.predict(new_students) # model한테 방금 만든 데이터로 합격 불합격을 예측(predict)시킨 결과를 results에 저장

for i in range(len(new_students)): # 학생 수만큼 반복
    study = new_students.iloc[i]["study_hours"] # 그 학생 데이터중에서 study_hours값만 가져오기
    attend = new_students.iloc[i]["attendance"] # 마찬가지로 attendance값만 가져오기
    result = results[i] # 학생의 합격/불합격 결과를 result에 저장하기

    if result == 1:
        print(f"공부 {study}시간, 출석 {attend}%인 학생은 합격입니다.") #result가 1일시 즉 합격일시 합격메세지출력
    else:
        print(f"공부 {study}시간, 출석 {attend}%인 학생은 불합격입니다.") #반대로 1이 아닐시 불합격메세지출력

# 문제:
# 각 학생의 로그를 출력하세요.
# 예시:
# 공부 ?시간, 출석 ?%인 학생은 합격 입니다. 
# 공부 ?시간, 출석 ?%인 학생은 불합격 입니다. 
```
